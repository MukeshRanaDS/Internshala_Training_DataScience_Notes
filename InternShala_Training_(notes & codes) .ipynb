# 8. T-Test

In statistics, t-tests are a type of hypothesis test that allows you to compare means. They are called t-tests because each t-test boils your sample data down to one number, the t-value.

A common analogy is that the t-value is the signal-to-noise ratio.

# 8.1 One-Sample t-test

    T-Test is conducted when we have very small sample and
    This test is to check whether the very small sample is the part of population or not.
    A 1-sample t-test compares one sample mean to a null hypothesis value.

```python
import pandas as pd
import scipy.stats as stats
from scipy.stats import ttest_1samp
```

```python
data = pd.read_csv('data_statistics.csv')
```

```python
data.head()
```

```python
data['Item_MRP'].mean()
```

```python
from scipy.stats import ttest_1samp
t_statistics, p_value = stats.ttest_1samp(data['Item_MRP'], 70)
```

```python
p_value
```

When the p-value is greater than 0.05, we fail to reject the null hypothesis(Sample mean = Population mean)

# 8.2 Two-Sample t-test :

    The 2-sample t-test takes our sample data from two groups and boils it down to the t-value. The process is very similar to the 1-sample t-test, and we can still use the analogy of the signal-to-noise ratio.
    The default null hypothesis for a 2-sample t-test is that the two groups are equal. We can see in the equation that when the two groups are equal, the difference (and the entire ratio) also equals zero. As the difference between the two groups grows in either a positive or negative direction, the signal becomes stronger.
    If we have 2 different sample & we have to check how difference these are from each other.

```python
import pandas as pd
import scipy.stats as stats
from scipy.stats import ttest_ind
```

```python
data = pd.read_csv('data_statistics.csv')
```

```python
data.head()
```

```python
data['Item_MRP'].mean()
```

```python
t_statistics, p_value = stats.ttest_ind(data['Item_MRP'], data['Item_Outlet_Sales'], equal_var=False)
```

```python
t_statistics
```

```python
p_value
```

As the p_value is less than 0.05, we can reject the hypothesis.

Null hypothesis means the 2 prices are similar(means are same) 

# 8.3 Paired t-test :

     The paired t-test and the 1-sample t-test are actually the same test in disguise!
    We check if the sample has changed it's behaviour after an intervation, or behaves significantly different in 2 conditions.

```python
import pandas as pd
import scipy.stats as stats
from scipy.stats import ttest_rel

data = pd.read_csv('data_statistics.csv')
data.head()
```

```python
t_statistics, p_value = stats.ttest_rel(data['Item_MRP'], data['Item_Outlet_Sales'])
```

```python
t_statistics
```

```python
p_value
```

As the p_value is less than 0.05, we can reject the null hypothesis(mean of Item_Outlet_Sales = mean of Item_MRP)

# 16. Unsupervised Learning

Hard Cluster -- When a point will lies in a cluster only.

Soft Cluster -- When a point may lies in more than one clusters.

k-means -- Its a Hard Clustering algorithm. k=2 means 2 clusters.

Iteration stopping criteria :
    
    1. Number of iterations
    
    2. If points doesnt changes their cluster number.
    
    3. Cluster iteration not changes(converse)

k-mean uses 'Eucliden Distance' and minimize it to assign an object to a cluster. 

Clustering does not have  a defined target variable. It is an unsupervised Learning algorithm.

For a given cluster size = 3, the resulting array from kmean.predict(data) can have values (0,1,2).

Optimized cluster value is selected using elbow curve, we use it to optimized number of cluster.
